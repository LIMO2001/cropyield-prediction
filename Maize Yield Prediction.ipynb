```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('maize_yield_dataset.csv')
```

```python
df.head()
```

```python
df_new=df[['Rainfall (mm)','Temperature (°C)','Soil Type','Soil pH','Soil Nutrients (N-P-K)','Pest & Disease Prevalence (%)','Previous Year Yield (kg/ha)','Maize Seed Variety']]
```

```python
df_new.head()
```

```python
df_new.rename(columns ={'Soil Nutrients (N-P-K)':'NPK','Pest & Disease Prevalence (%)':'Pest Prevalance','Previous Year Yield (kg/ha)':'Yield','Maize Seed Variety':'Variety'})
```

```python
df_new.describe()
```

```python
df_new.isnull().sum()
```

```python
df_new.isna().sum()
```

```python
df_new.loc[df_new['Soil Type'] == 'Clay', 'Soil Type'] = 1
```

```python
df_new.loc[df_new['Soil Type'] == 'Loam', 'Soil Type'] = 2
df_new.loc[df_new['Soil Type'] == 'Sandy', 'Soil Type'] = 3
df_new.head()
```

```python
df_new.rename(columns ={'Soil Nutrients (N-P-K)':'NPK','Pest & Disease Prevalence (%)':'Pest Prevalance','Previous Year Yield (kg/ha)':'Yield','Maize Seed Variety':'Variety'},inplace=True)
```

```python
df_new.head()
```

```python
df_new[['N', 'P', 'K']] = df_new['NPK'].str.split('-', expand=True)
```

```python
df_new.head()
```

```python
df_new.drop(labels=['NPK'],inplace=True,axis=1)
```

```python
df_new.head()
```

```python
df_new.tail()
```

```python
df_new['Variety'].unique()
```

```python
df_new.loc[df_new['Variety'] == 'Hybrid-1', 'Variety'] = 1
df_new.loc[df_new['Variety'] == 'Hybrid-2', 'Variety'] = 3
df_new.loc[df_new['Variety'] == 'Local-1', 'Variety'] = 2
df_new.head()
```

```python
df_new.corr()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled= scaler.fit_transform(df_new)
df_scaled = pd.DataFrame(df_scaled, columns=df_new.columns)
print(df_scaled)
```

```python
df_scaled.corr()
```

```python
plt.scatter(df_scaled['Rainfall (mm)'],df_scaled['Yield'],c="blue")
plt.scatter(df_scaled['Temperature (°C)'],df_scaled['Yield'],c="red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Rainfall and Temperature against Yield")
plt.show()
```

```python
import seaborn as sns
sns.violinplot(x='Rainfall (mm)', data=df_scaled, hue='Rainfall (mm)')

# Show the plot
plt.show()
```

```python
df_scaled.corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_scaled.drop(labels=['Yield'],axis=1)
```

```python
y=df_scaled['Yield']
x.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
regression_model=LinearRegression()
```

```python
regression_model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred=regression_model.predict(X_test)
print(f"The MSE is ",mean_squared_error(y_pred,Y_test))
```

```python
plt.scatter(y_pred,Y_test,c="blue")
```

```python
import joblib
joblib.dump(regression_model,"model.pkl")
```
